```python
## Basic Statistics and Distribution
# 1. What are the mean, median, and standard deviation of the HP, Attack, Defense, Sp. Atk, Sp. Def, and Speed attributes?

# mean of the HP, Attack, Defense, Sp. Atk, Sp. Def, and Speed attributes in a DataFrame(test_data_1)
test_data_1.loc[:,"HP":"Speed"].mean()
```

```python
# median of the HP, Attack, Defense, Sp. Atk, Sp. Def, and Speed attributes in a DataFrame(test_data_1)
test_data_1.loc[:,"HP":"Speed"].median()
```

```python
# std of the HP, Attack, Defense, Sp. Atk, Sp. Def, and Speed attributes in a DataFrame(test_data_1)
test_data_1.loc[:,"HP":"Speed"].std()
```

```python
# Bar Graph
plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
x1 = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
y1 = test_data_1.loc[:,"HP":"Speed"].mean()
plt.bar(x1, y1, label='First Bars',color="blue")
plt.xlabel('Category')
plt.ylabel('values')
plt.title('Mean Bar Graph')
plt.legend()

plt.subplot(1,3,2)
x2 = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
y2 = test_data_1.loc[:,"HP":"Speed"].median()
plt.bar(x2, y2, label='Second Bars',color="green")
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Median Bar Graph')
plt.legend()

plt.subplot(1,3,3)
x3 = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
y3 = test_data_1.loc[:,"HP":"Speed"].std()
plt.bar(x3, y3, label='Third Bars',color="orange")
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Standard Deviation Bar Graph')
plt.legend()

plt.show()
```

```python
# type, category, value
# mean, hp, 69.258750
# mean, attack, 79.001250
```

```python
# Line Graph
x = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
y1 = test_data_1.loc[:,"HP":"Speed"].mean()
y2 = test_data_1.loc[:,"HP":"Speed"].median()
y3 = test_data_1.loc[:,"HP":"Speed"].std()

plt.figure(figsize=(10,4))
plt.plot(x,y1,color="blue",label = "Mean", marker='s', linestyle='dashdot')
plt.plot(x,y2,color="yellow" ,label = "Median", marker='o', linestyle='dashed')
plt.plot(x,y3,color="green" ,label = "Standard Deviation", marker='d')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")
plt.legend()
plt.show()
```
